### Строим график

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np 
x = np.linspace(-6, 6, 100)
y = np.sin(x)
plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title('Пример sin')
plt.xlabel('время')
plt.ylabel('давление')
plt.show()
```

```python
x = np.linspace(-6, 6, 100)
y = np.cos(x)
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="Sin wave")
plt.title('Пример cos')
plt.xlabel('время')
plt.ylabel('давление')
plt.legend()
plt.show()
```
